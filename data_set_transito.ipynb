```python
Q1 = ds['pavement_quality'].quantile(0.25)
Q3 = ds['pavement_quality'].quantile(0.75)
IQR = Q3 - Q1

limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

ds_sem_outliers = ds[(ds['pavement_quality'] >= limite_inferior) & (ds['pavement_quality'] <= limite_superior)]

ds_sem_outliers.boxplot(column="pavement_quality")
plt.show()
```
