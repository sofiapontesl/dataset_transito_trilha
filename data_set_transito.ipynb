```python
Q1 = ds['rain_intensity'].quantile(0.25)
Q3 = ds['rain_intensity'].quantile(0.75)
IQR = Q3 - Q1

limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

ds_sem_outliers = ds[(ds['rain_intensity'] >= limite_inferior) & (ds['rain_intensity'] <= limite_superior)]

ds_sem_outliers.boxplot(column="rain_intensity")
plt.show()
```
